```python

print("-------")



with open ('../data/topics_definitions.txt', 'rb') as fp:
    topics_documents = pickle.load(fp)
with open('../data/topics_details.txt', 'rb') as fp2:
    topics_info = pickle.load(fp2)
with open('../data/topics_uris.txt', 'rb') as fp3:
    topics_uris = pickle.load(fp3)

#The following is hardcoded for the 7th edition
indices_ed=["7"]

terms_per_volume = [102, 82, 150, 77, 104, 141, 89, 85, 90, 109,
                    91, 106, 73, 98, 93, 92, 83, 101, 111, 146]
#this calculates what the indicies are for each volume
indicies = []
indicies.append(0)
for i in range(len(terms_per_volume)):
    indicies.append(sum(terms_per_volume[:i+1]))

print("Phase 1")
model = TransformerSummarizer(transformer_type="XLNet",transformer_model_key="xlnet-base-cased")
print("Phase 2")
topics_summary=[]
cont=0
for vol in range(len(indicies)-1):
    print("VOL ", vol + 2)
    for i in range(indicies[vol], indicies[vol+1]):
        #print("Summary %s-%s" %("7", i))
        text=topics_documents[i]
        try:
            full = ''.join(model(text, min_length=60, max_length=300))
            topics_summary.append(full)
        except:
            print("Error in %s" %cont)
            topics_summary.append(["Not avaiable"])
    cont=cont+1
    
    print("FINISH TOPICS FOR EDITION 7 VOL %s"%(vol+2))
    

print("FINISH TOPICS FOR EDITION 7")
filename='../data/topics_summary_7.txt'
with open(filename, 'wb') as fp:
    pickle.dump(topics_summary, fp)
```

```python
with open('../data/topics_uris.txt', 'rb') as fp:
    topics_uris = pickle.load(fp)
with open ('../data/terms_definitions.txt', 'rb') as fp2:
    documents = pickle.load(fp2)
with open('../data/terms_uris.txt', 'rb') as fp3:
    uris = pickle.load(fp3)

indices_ed=["7"]
terms_per_volume = [102, 82, 150, 77, 104, 141, 89, 85, 90, 109,
                    91, 106, 73, 98, 93, 92, 83, 101, 111, 146]
#this calculates what the indicies are for each volume
indicies = []
indicies.append(0)
for i in range(len(terms_per_volume)):
    indicies.append(sum(terms_per_volume[:i+1]))

print("Phase 2")
for ed in indices_ed:
    print("EDITION %s" %ed)
    cont=0
    filename_t='../data/topics_summary_7.txt'
    with open(filename_t, 'rb') as fp5:
        topics_summary= pickle.load(fp5)
    for i in indicies:
        uri=topics_uris[i]
        index_i=uris.index(uri)
        documents[index_i]=topics_summary[cont]
        cont=cont+1
    
print("FINISH")
with open('../data/topics_summaries_total.txt', 'wb') as fp:
    pickle.dump(documents, fp)

print("SAVED")
```
